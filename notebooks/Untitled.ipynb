```python
import tensorflow as tf
```

```python
import pandas as pd
import numpy as np
```

```python
train_raw = pd.read_csv("../data/raw/train.csv")
test_raw = pd.read_csv("../data/raw/test.csv")
```

```python
train_raw
```

```python
train_raw.dtypes
```

```python
feature_list = ["Pclass", "Age", "SibSp", "Parch", "Fare"]
```

```python
train_preprosses = train_raw[["PassengerId", "Survived"] + feature_list]
```

```python
x_train = train_preprosses[feature_list].fillna(0).values
y_train = train_preprosses[["Survived"]].values.flatten()

x_test = test_raw[feature_list].values
```

```python
x_train
```

```python
y_train
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dense(8, activation='relu'),
  tf.keras.layers.Dense(2, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=500)
```

```python
y_test = model.predict_classes(x_test)
```

```python
test_result = pd.DataFrame({'PassengerId' : test_raw["PassengerId"], 'Survived' : y_test})
```

```python
test_result
```

```python
test_result.to_csv("../data/processed/result_2.csv", index=False)
```
